# Pokemon Clustering
Given a set of pokemon, I want to cluster them based on color and other features that my be similar using K-Meanss

```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

parent_path = "../input/pokemon-images-dataset/pokemon/pokemon/"

images = []
for image_path in os.listdir(parent_path):
    url = parent_path + image_path
    im = Image.open(url)
    data = np.asarray(im).reshape(-1)
    images.append(data)
    
dataset = np.array(images)

def show_image(index):
    plt.imshow(dataset[index].reshape(256, 256, 4))
    
scaler = StandardScaler()
scaled_dataset = scaler.fit_transform(dataset)

VARIANCE = 0.95
pca = PCA(VARIANCE)

compressed_dataset = pca.fit_transform(scaled_dataset)
print(compressed_dataset.shape)
```

```python
from sklearn.cluster import KMeans

NUM_CLUSTERS = 17
kmeans = KMeans(n_clusters = NUM_CLUSTERS, init = "k-means++", n_init = 10).fit(compressed_dataset)
pred = kmeans.predict(compressed_dataset)

cluster = 9
rows = 40
cols = 10
axes=[]
fig=plt.figure(figsize = (100, 100))
count = 1
for i in range(len(pred)):
    if pred[i] == cluster:
        x = dataset[i].reshape(256, 256, 4)
        axes.append(fig.add_subplot(rows, cols, count))
        plt.imshow(x)
        count += 1
fig.tight_layout()    
plt.show()
```
